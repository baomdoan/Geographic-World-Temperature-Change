```python
def convert(data):
    # first convert country name to its equivalent alpha 2 code
    cn_code = pc.country_name_to_country_alpha2(data.Area, cn_name_format="default")
    if cn_code == 'AQ':
        return 'AQ'
    elif cn_code == 'TF':
        return 'TF'
    elif cn_code == 'VA':
        return 'VA'
    elif cn_code == 'PN':
        return 'PN'
    elif cn_code == 'TL':
        return 'TL'
    elif cn_code == 'UM':
        return 'UM'
    elif cn_code == 'EH':
        return 'EH'
    # convert alpha2 to continent
    conti_code = pc.country_alpha2_to_continent_code(cn_code)
    return conti_code
```

```python
df_filtered_countries['Continent']= df_filtered_countries.apply(convert, axis=1)
df_filtered_countries
```

```python
conti_names = {'AS':'Asia',
               'SA':'South America',
               'OC':'Oceania',
               'EU':'Europe',
               'NA':'North America',
               'AF':'Africa'}
df_filtered_countries['Continent']=df_filtered_countries['Continent'].map(conti_names)
```
